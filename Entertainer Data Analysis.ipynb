# Univariate Analysis

```python
def numerical_univariate_analysis(numerical_data) :
     for col_name in numerical_data:
            print("*"*10, col_name,"*"*10)
            print(numerical_data[col_name].agg(['min','max']))
            print()
```

```python
numerical_univariate_analysis(Entertainer_data[['Year of First Oscar']])
```

```python
numerical_univariate_analysis(Entertainer_data[['Year of Breakthrough']])
```
